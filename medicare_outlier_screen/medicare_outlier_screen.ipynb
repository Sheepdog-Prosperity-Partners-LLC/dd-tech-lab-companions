# Medicare Outlier Screen

This notebook runs the paired Python script against a synthetic CMS-shaped cache. The output is a lead list, not a finding.


```python
from pathlib import Path
import subprocess
import sys

folder = Path.cwd()
if not (folder / "medicare_outlier_screen.py").exists():
    folder = Path.cwd() / "medicare_outlier_screen"
out = Path("/tmp/medicare_outlier_synthetic.csv")
cmd = [
    sys.executable,
    str(folder / "medicare_outlier_screen.py"),
    "--hcpcs", "36902",
    "--cache", str(folder / "data/sample/medicare_hcpcs_36902_synthetic.csv"),
    "--min-peers", "5",
    "--z", "2.0",
    "--out", str(out),
]
result = subprocess.run(cmd, check=True, text=True, capture_output=True)
print(result.stdout)
print(result.stderr)
print("Output CSV:", out)
```
