# D1 Integrated Workflow

Combine synthetic intake flags into a lead-only diligence triage row for the integrated workflow.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build D1 integrated workflow output')
run_command([sys.executable, '-m', 'ns_diligence.risk_workflow', 'data/synthetic/d1_deal_intake_flags.csv', '/tmp/ns-diligence-lab-notebooks/d1_integrated_workflow_sample.csv'])
```
