# A2 Outbound Triage

Run the synthetic outbound-investment triage lablet. Outputs are no obvious trigger, review, or escalate leads only.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build A2 outbound triage sample output')
run_command([sys.executable, '-m', 'ns_diligence.outbound_triage', 'data/synthetic/a2_outbound_targets.csv', '/tmp/ns-diligence-lab-notebooks/a2_outbound_triage_sample.csv'])
```
