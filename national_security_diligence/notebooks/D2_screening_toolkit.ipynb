# D2 Screening Toolkit

Run the synthetic capstone screens: name screening, export-list screening, and ownership teaching graph. Outputs are leads only.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build D2 synthetic name-screen output')
run_command([sys.executable, 'ns_screen.py', 'data/synthetic/d2_counterparties.csv', 'data/sample/d2_fixture_screening_list.csv', '/tmp/ns-diligence-lab-notebooks/d2_ns_screen_sample.csv', '--threshold', '0.80'])
```

```python
print('Build D2 export-list teaching output')
run_command([sys.executable, '-m', 'ns_diligence.export_list_screen', 'data/synthetic/d2_counterparties.csv', 'data/sample/d2_fixture_screening_list.csv', '/tmp/ns-diligence-lab-notebooks/d2_export_screen_sample.csv', '--threshold', '0.80'])
```

```python
print('Build D2 synthetic ownership output')
run_command([sys.executable, '-m', 'ns_diligence.ownership_graph', 'data/synthetic/d2_ownership_edges.csv', '/tmp/ns-diligence-lab-notebooks/d2_ownership_graph_sample.csv'])
```
