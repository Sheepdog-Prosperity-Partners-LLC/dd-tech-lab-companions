# C2 Forfeiture Timeline Builder

Build a public-release forfeiture timeline table from the committed sample events. The rows are timeline aids, not findings.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build C2 forfeiture timeline output')
run_command([sys.executable, '-m', 'ns_diligence.timeline_builder', 'data/sample/c2_public_forfeiture_events.csv', '/tmp/ns-diligence-lab-notebooks/c2_forfeiture_timeline_sample.csv'])
```
