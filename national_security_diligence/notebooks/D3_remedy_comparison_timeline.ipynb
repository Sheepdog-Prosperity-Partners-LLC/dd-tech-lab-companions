# D3 Remedy Comparison Timeline

Build the publication-approved public remedy-comparison timeline from committed sample events.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build D3 remedy comparison output')
run_command([sys.executable, '-m', 'ns_diligence.remedy_comparison', 'data/sample/d3_remedy_comparison_events.csv', '/tmp/ns-diligence-lab-notebooks/d3_remedy_comparison_timeline.csv'])
```
