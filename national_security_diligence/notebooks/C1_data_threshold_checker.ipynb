# C1 Data Threshold Checker

Run the EO 14117 synthetic data threshold checker. The code flags threshold issues for follow-up and does not make a legal conclusion.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build C1 data threshold sample output')
run_command([sys.executable, '-m', 'ns_diligence.data_threshold_checker', 'data/synthetic/c1_data_assets.csv', '/tmp/ns-diligence-lab-notebooks/c1_data_threshold_sample.csv'])
```
