# C3 Bounty Program Matrix

Build the whistleblower and bounty program comparison matrix from public program terms.


```python
from pathlib import Path
import os
import subprocess
import sys

root = Path.cwd()
if not (root / "src").exists():
    root = Path.cwd().parent
if not (root / "src").exists():
    raise FileNotFoundError("Run this notebook from the national_security_diligence folder or its notebooks folder.")

scratch = Path("/tmp/ns-diligence-lab-notebooks")
scratch.mkdir(parents=True, exist_ok=True)
env = os.environ.copy()
env["PYTHONPATH"] = str(root / "src")

def run_command(args):
    completed = subprocess.run(args, cwd=root, env=env, text=True, capture_output=True, check=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.stderr:
        print(completed.stderr)
```

```python
print('Build C3 bounty program matrix output')
run_command([sys.executable, '-m', 'ns_diligence.award_matrix', 'data/sample/c3_whistleblower_programs.csv', '/tmp/ns-diligence-lab-notebooks/c3_bounty_program_matrix.csv'])
```
