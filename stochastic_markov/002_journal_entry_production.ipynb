# 002 Journal Entry Production

This notebook runs the paired Python script as the canonical source for the worked example. The script sets deterministic seeds where needed and prints the article workpaper output.


```python
from pathlib import Path
import os
import runpy

script_name = "002_journal_entry_production.py"
candidate_dirs = [
    Path.cwd(),
    Path.cwd() / "stochastic_markov",
    Path.cwd().parent / "stochastic_markov",
]
script_path = next((directory / script_name for directory in candidate_dirs if (directory / script_name).exists()), None)
if script_path is None:
    raise FileNotFoundError(f"Could not find {script_name} from {Path.cwd()}")

old_cwd = Path.cwd()
os.chdir(script_path.parent)
try:
    runpy.run_path(str(script_path), run_name="__main__")
finally:
    os.chdir(old_cwd)
```
